### Data Cleaning <a class="anchor" id="cleaning"> </a>
There are clearly no outliers in any of the numerical data, as can be seen from the boxplots.

There are also no spelling mistakes in the remaining text data.

```python
# Check for outliers in numerical data
num_column_names = np.concatenate((spotify_data.columns[4:9], spotify_data.columns[10:12]))
# print(num_column_names)
fig, axs = plt.subplots(1, len(num_column_names), figsize=(20,5))
for i in range(len(num_column_names)):
    axs[i].boxplot(spotify_data[num_column_names[i]])
    axs[i].set_title(num_column_names[i])
    axs[i].set_xticks([])
plt.tight_layout()
plt.show()
```

```python
# Check how many countries, how many genres, how many platform types
no_num_column_names = np.concatenate((spotify_data.columns[0:4], spotify_data.columns[9:10]))
for i in range(len(no_num_column_names)):
    print(no_num_column_names[i] + ": ", spotify_data[no_num_column_names[i]].nunique())
    print(spotify_data[no_num_column_names[i]].unique())
```

### Exploratory Data Analysis <a class="anchor" id="EDA"></a>
1. Overview of Global Streaming Trends
    - Goal: Understand the overall landscape of Spotify streaming data.
    - Questions:
        - Which genres are the most streamed globally?
        - Are newer albums streamed more or less than older ones?
        - What are the top artists by total streams?

2. Geographic Listening Patterns
    - Goal: Explore how music consumption patterns vary across countries.
    - Questions:
        - Which countries contribute the most to global streams?
        - How do skip rates vary across countries?
        - Are certain countries more inclined toward specific genres?

3. Artist Spotlights & Performance
    - Goal: Investigate how different artists perform across streaming metrics.
    - Questions:
        - Which artists have the highest number of monthly listeners?
        - Which artists generate the most total hours streamed?

4. Platform Behavior Analysis
    - Goal: Compare user behavior between Free and Premium platforms.
    - Questions:
        - How do skip rates differ between Free and Premium users?
        - Do Free or Premium users stream for longer durations?
        - Is there a difference in streams per album by platform?

5. Genre-Based Streaming Insights
    - Goal: Identify behavioural trends patterns across music genres.
    - Questions:
        - How do average stream durations differ by genre?
        - Do certain genres have higher skip rates than others?
